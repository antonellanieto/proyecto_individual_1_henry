```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast  
```

```python
df = pd.read_csv('steam_games.csv')
df
```

```python
df['genres'] = df['genres'].astype(str)
df['genres'] = df['genres'].str.extract(r"(\w+)")
df
```

```python
df['tags'] = df['tags'].astype(str)
df['tags'] = df['tags'].str.extract(r"(\w+)")
df
```

```python
df['specs'] = df['specs'].astype(str)
df['specs'] = df['specs'].str.extract(r"(\w+)")
df['sentiment'] = df['sentiment'].astype(str)
df['sentiment'] = df['sentiment'].str.extract(r"(\w+)")
df
```

```python
df['genres'].fillna('Unknown', inplace=True)
df['specs'].fillna('Unknown', inplace=True)
df['sentiment'].fillna('Unknown', inplace=True)
df['publisher'].fillna('Unknown', inplace=True)
df['genres'].fillna('Unknown', inplace=True)
df['app_name'].fillna('Unknown', inplace=True)
df['title'].fillna('Unknown', inplace=True)
df['url'].fillna('Unknown', inplace=True)
df['tags'].fillna('Unknown', inplace=True)
df['reviews_url'].fillna('Unknown', inplace=True)
df['developer'].fillna('Unknown', inplace=True)
df['release_date'] = df['release_date'].fillna('1999-12-25')
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_date'] = df['release_date'].fillna(pd.to_datetime('2000-01-01'))

df
```

```python
df['genres'] = df['genres'].astype(str)
df['genres'] = df['genres'].str.extract(r"(\w+)")
df['discount_price'].fillna(0, inplace=True)
df['price'].replace("Free to play", 0, inplace=True)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price'].fillna(0, inplace=True)
df['metascore'] = pd.to_numeric(df['metascore'], errors='coerce')
df['metascore'].fillna(0, inplace=True)
df['release_date'] = df['release_date'].fillna('1999-12-25')
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df.dropna(subset=['release_date'])

df
```

```python
from sklearn.preprocessing import LabelEncoder

columns_to_encode = ['publisher', 'app_name', 'title', 'url', 'developer', 'sentiment', 'genres', 'tags', 'reviews_url', 'developer', 'specs', 'sentiment']
label_encoder = LabelEncoder()
for column in columns_to_encode:
    df[column + '_encoded'] = label_encoder.fit_transform(df[column])


df
```

```python
correlation = df['genres_encoded'].corr(df['price'])

plt.figure(figsize=(8, 6))
plt.scatter(df['genres_encoded'], df['price'])
plt.title("Scatter Plot of genres encoded vs Price")
plt.xlabel("genre_encoded")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
metascore = 'metascore'
# Calcular el primer y tercer cuartil
Q1 = df[metascore].quantile(0.25)
Q3 = df[metascore].quantile(0.75)
# Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1
# Definir umbrales para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Identificar outliers en la columna especificada
outliers = df[(df[metascore] < lower_bound) | (df[metascore] > upper_bound)]

outliers['metascore'].sum()
```

```python
df = df.drop(outliers.index)
```
